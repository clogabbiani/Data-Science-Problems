# Sales Analysis

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

Import Data

#### Task #1: Merging 12 months of sales data into a single file

```python
df_april = pd.read_csv("./Sales_Data/Sales_April_2019.csv")
df_april.head()
```

```python
# Let's load all files

path = "./Sales_Data"

files = [file for file in os.listdir(path)]

for file in files:
    print(file)
```

```python
# concatenate files into a single csv

all_months_data = pd.DataFrame()

for file in files:
    df = pd.read_csv("./Sales_Data/" + file)
    all_months_data = pd.concat([all_months_data,df])
    
all_months_data.head()

all_months_data.to_csv("all_data.csv", index=False)
```

#### Read in updated dataframe

```python
all_data = pd.read_csv("all_data.csv")
all_data.head()
```

```python
# check this strange issue
all_data[all_data['Quantity Ordered'] == 'Quantity Ordered']
```

```python
all_data = all_data[all_data['Quantity Ordered'] != 'Quantity Ordered']
all_data.shape
```

### Clean up the data

```python
# drop nan rows
all_data = all_data.dropna(how='any')
all_data.head()
```

```python
all_data.shape
```

```python
# Convert columns to the correct type

all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
```

#### Augment data with additional columns

#### Task #2: add month column

```python
all_data['Order Date'].iloc[0][:2]
```

```python
all_data['Month'] = all_data['Order Date'].str[:2]
all_data = all_data[all_data['Month'].str[:2] != 'Or']
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

#### Task #3: Add sales column

```python
all_data['Sales'] = all_data['Quantity Ordered']*all_data['Price Each']
all_data.head()
```

#### Question 1: What is the best month for sales? How much was earnerd that month?

```python
all_data.groupby('Month').sum()
```

```python
# Best month is December! Worst January
```

```python
# PLOT

months = range(1,13)
results = all_data.groupby('Month').sum()['Sales']

plt.bar(months, results)
plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Month number')
plt.show()
```

#### Question 2: What city has the highest number of sales?

```python
# lets add the city column to the df

def get_city(address):
    return address.split(',')[1][1:]

def get_state(address):
    return address.split(',')[2].split(' ')[1]

all_data['City'] = all_data['Purchase Address'].apply(lambda x: f"{get_city(x)} {get_state(x)}")

all_data.head()
```

```python
results = all_data.groupby('City').sum()
results
```

```python
# PLOT

cities = [city for city, df in all_data.groupby('City')]

plt.bar(cities, results['Sales'])
plt.xticks(cities, rotation='vertical', size=8)
plt.ylabel('Sales in USD ($)')
plt.xlabel('City name')
plt.show()
```

#### Question 3: What time is more likely to sell the product?

```python
# convert 'Order Date' into a DateTime object
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
all_data.head()
```

```python
def hr_func(ts):
    return ts.hour

all_data['Hour'] = all_data['Order Date'].apply(hr_func)

all_data['Minute'] = all_data['Order Date'].dt.minute

all_data.head()
```

```python
# PLOT

hours = [hour for hour, df in all_data.groupby('Hour')]

plt.plot(hours, all_data.groupby('Hour').count())
plt.xticks(hours)
plt.xlabel('Hours')
plt.ylabel('Number of Orders')
plt.grid()
plt.show()
```

```python
# we can do the same for each city
```

```python
cities = [city for city, df in all_data.groupby('City')]
for city in cities:
    data_city = all_data[all_data['City'] == city]
    plt.plot(hours, data_city.groupby('Hour').count())
    plt.xticks(hours)
    plt.title('City of ' + city)
    plt.xlabel('Hours')
    plt.ylabel('Number of Orders')
    plt.grid()
    plt.show()
```

#### Question 4: What products are most often sold together?

```python
all_data.head()
```

```python
# count all duplicates using the Order ID, mantainig only those rows
df = all_data[all_data['Order ID'].duplicated(keep=False)]
df.head(30)

# add a column to list all toghether products
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))

# remove 
df = df[['Order ID', 'Grouped']].drop_duplicates()
```

```python
df.head(10)
```

```python
# References: 
from itertools import combinations
from collections import Counter
```

```python
count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))
    
for key,value in count.most_common(10):
    print(key, value)
```

```python
# ('iPhone', 'Lightning Charging Cable')  is the most popular tandem objects
```

#### Question 5: What product sold the most? 

```python
df_prod = all_data.groupby(['Product'])
df_prod.head(30)
```

```python
# PLOT

quantity_ordered = df_prod.sum()['Quantity Ordered']

products = [product for product, df in df_prod]

plt.bar(products, quantity_ordered)
plt.xticks(products, rotation='vertical', size=8)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Product name')
plt.show()
```

```python
prices = all_data.groupby('Product').mean()['Price Each']

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered, color='green')
ax2.plot(products, prices, 'r-')

ax1.set_xlabel('Poduct Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price ($)', color='b')
ax1.set_xticklabels(products, rotation='vertical', size=8)

plt.show()
```
